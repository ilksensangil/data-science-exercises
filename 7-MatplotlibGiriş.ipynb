```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi=[10,20,30,30,30,40,50,60,70,75]
```

```python
kiloListesi=[20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi=np.array(yasListesi) # numpya çevirdik.
```

```python
numpyKiloListesi=np.array(kiloListesi)
```

```python
numpyYasListesi
```

```python
numpyKiloListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"g") 
# x ekseninde yaş y ekseninde kilo var g dediğimiz rengi yeşil renk ile gösterdik.
plt.xlabel("yas") # x ekseninin etiketi
plt.ylabel("kilo") # y ekseninin etiketi
plt.title("kilonun yaşa göre grafiği") # grafiğin başlığı
```

```python
numpyDizisi1=np.linspace(0,10,20)  # 0 ile 10 arasında 20 tane değişken oluştur
```

```python
numpyDizisi1
```

```python
numpyDizisi2=numpyDizisi1**3  # hepsinin üç üssünü  al
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g") # grafiğini çizdirdik.
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g--") # kesikli çizgi oluşturdu.
```

```python
plt.subplot(1,2,1) # alt grafiğini çizdik.1. parametre kaç sıra olduğu
                                        # 2. parametre kaç grafik olduğu
                                        # 3. parametre ise ilk önce hangisini göstereceğimiz
plt.plot(numpyDizisi1,numpyDizisi2,"r--")
plt.subplot(1,2,2)                      # 2.grafiği gösterdik
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
benfigure=plt.figure() # boş bir çerçeve oluşturduk
```

```python
benfigure=plt.figure()
figureAxes=benfigure.add_axes([0.2,0.2,0.9,0.9])  # grafiğin genişlik ve uzunluk değerleri
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X ekseni veri ismi")
figureAxes.set_ylabel("y ekseni veri ismi")
figureAxes.set_title("grafik başlığı")
```

```python
figure2=plt.figure()

eksen1=figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2=figure2.add_axes([0.3,0.3,0.3,0.3])
eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X ekseni veri ismi")
eksen1.set_ylabel("y ekseni veri ismi")
eksen1.set_title("ana grafik başlığı")
eksen2.plot(numpyDizisi1,numpyDizisi2,"g")
eksen2.set_xlabel("X ekseni veri ismi")
eksen2.set_ylabel("y ekseni veri ismi")
eksen2.set_title("küçük grafik başlığı")
```

```python
figure2=plt.figure()

eksen1=figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2=figure2.add_axes([0.1,0.4,0.3,0.3]) # 0.3 kısmını değiştiridik küçük grafik sola kaydı 0.4 olan kısım ise yukarı çıktı
eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X ekseni veri ismi")
eksen1.set_ylabel("y ekseni veri ismi")
eksen1.set_title("ana grafik başlığı")
eksen2.plot(numpyDizisi1,numpyDizisi2,"g")
eksen2.set_xlabel("X ekseni veri ismi")
eksen2.set_ylabel("y ekseni veri ismi")
eksen2.set_title("küçük grafik başlığı")
```

```python
(benimfigure,benimeksenler)=plt.subplots(nrows=1,ncols=2) # bunu kullanarak iç içe grafik yapabiliriz
for eksen in benimeksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X ekseni")
plt.tight_layout()  # grafiklerin birbirine girmesini önler
```

```python
yeniFigure=plt.figure()
yeniEksen=yeniFigure.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi2**2,label="numpyDizisi**2")  # etiket kısmını grafikte gösterdik.
yeniEksen.plot(numpyDizisi1,numpyDizisi1**3,label="numpyDizisi**3") # ikinci etikette gösterildi.
yeniEksen.legend() # matplotlib ikisine de ayrı renk koydu.
```

```python
yeniFigure=plt.figure(figsize=(3,3)) # grafiğin boyutunu değiştirdi
yeniEksen=yeniFigure.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi2**2,label="numpyDizisi**2")  
yeniEksen.plot(numpyDizisi1,numpyDizisi1**3,label="numpyDizisi**3") 
yeniEksen.legend(loc=1) # legend kısmı sağa gitti
```

```python
yeniFigure=plt.figure(dpi=100) # grafiğin boyutunu değiştirdi daha kaliteli görüntü verdi.
yeniEksen=yeniFigure.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi2**2,label="numpyDizisi**2")  
yeniEksen.plot(numpyDizisi1,numpyDizisi1**3,label="numpyDizisi**3") 
yeniEksen.legend(loc=1) 
```

```python
yeniFigure.savefig("benimfigur.png") # grafiğin görüntüsünü kaydettik
```
