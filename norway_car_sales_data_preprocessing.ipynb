models.

monthly total sales.

```python
import numpy as np
df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'],
'key2' : ['one', 'two', 'one', 'two', 'one'],
'data1' : np.random.randn(5),
'data2' : np.random.randn(5)})
```

```python
df
```

```python
df.groupby('key1').mean() # key1 e göre değişkenlerin ortalamasını grupla
```

that there is no key2 column in the result. Because df['key2'] is not numeric data

if you use more than one column for grouping, resulting dataframe have multiindex format

```python
df.groupby(['key1',"key2"]).mean() # tabloyu key1 ve key2 göre grupla 
```

let'select only data1 column to get the mean

```python
df.groupby(['key1',"key2"])["data1"].mean() # data1'in ortalamasını aldı.
```

Depending on your selection style you may get the result in dataframe or series format

```python
df.groupby(['key1',"key2"])[["data1"]].mean() # köşeli parantez bize tabloyu getirdi.
```
